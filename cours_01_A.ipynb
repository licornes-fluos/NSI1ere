Toute **machine** informatique manipule une représentation des **données** dont **l’unité minimale est le bit 0/1**, ce qui permet d’unifier logique et calcul. Le codage **conditionne la taille des différentes valeurs en mémoire**.

### II. Conservation des données
**Avant l'ère de la numérisation massive** qui date du début des années 2000, la manière de conserver les données **dépendait du type des données** :
* Les écrits dans des livres.
* Les images sur du papier-photo.
* Les sons sur des bandes magnétiques ou des disques vinyles.
* Les vidéos sur des films.

**Aujourd'hui**, toutes ces données cohabitent dans des ordinateurs sous forme de **fichiers dont seule l'extension (après le ".") permet de connaître la nature** :
* info.txt ou info.doc ou encore info.odt contiennent du texte.
* info.jpg ou info.bmp ou encore info.png contiennent des images.
* info.mp3 ou info. wav contiennent des sons.
* info.avi ou info.divx contiennent des vidéos.

### III. Numérisation de l'information
Dans un **ordinateur**, toute **information**, aussi complexe soit elle, est **codée sous la forme d'une suite de 0 et de 1**. Le bit (Binary digiT) est donc la **quantité élémentaire d'information** (la plus petite information), il ne peut prendre que deux valeurs : 0 ou 1 (ou *vrai/faux* ou *oui/non* ou *haut/bas* ...).

On peut **démontrer mathématiquement** que n'importe quelle information, quelle que soit sa nature (texte, image, audio, ...) peut être codée de cette façon.

Sans faire de démonstration détaillée, admettons que **n'importe quel nombre entier positif peut être codé avec un nombre binaire** (suite de 0 et de 1) **suffisamment grand**, alors on peut coder :
* les **entiers négatifs** : il suffit d'ajouter à la fin un bit valant 0 si le nombre est négatif et 1 s'il est positif,
* les **nombres réels** : il suffit de coder séparément la partie entière et la partie décimale qui sont toutes les deux des nombres entiers (ajouter un bit pour le signe),
* les **caractères** : on associe un nombre entier, et donc un nombre binaire, à chaque caractère,

* un **texte** : c'est une suite de caractères, donc de nombres entiers, donc de nombres binaires,
* une **image** : on associe un nombre entier (ou 3 si c'est en couleurs), codé en binaire, à chaque pixel pour indiquer sa luminosité,
* un **son** : c'est un signal temporel (tension électrique fournie par le micro en fonction du temps) donc une suite de nombres qui peuvent être codés en binaire,
* une **vidéo** : c'est une suite d'images avec du son ...


### IV. Histoire du codage binaire
Georges **Boole**, en 1847, invente **l'algèbre booléenne** qui permet de traduire mathématiquement les raisonnements logiques à partir des opérateurs logiques *ET* et *OU* appliqués aux valeurs *0* et *1* (ou *Vrai* et *Faux*).

En 1937, Claude **Shanonn** montre que l'algèbre boléenne peut s'appliquer à des **circuits électroniques** où le 1 correspond au passage d'un courant électrique et le 0 à l'absence de courant.

A partir de là, on peut commencer à construire des **machines qui manipulent des 0 et des 1 pour faire des opérations logiques**, c'est le début de **l'informatique moderne**. Plus tard, la **miniaturisation** des transistors (composant de base en électronique) permettra le développement de la **micro-informatique**.

**Remarque :** Le 0 et le 1 ne sont que des **représentations symboliques** de l'état d'une information élémentaire, dans un micro-processeur cela se traduit par le passage ou non d'un courant mais sur un disque dur cela se traduit par l'orientation d'une sorte de minuscule aimant et sur un CD-ROM cela se traduit par la présence ou non d'un trou (gravure).


### V. Compter la quantité d'informations
On a vu que le **bit** (symbole "b" parfois utilisé) consitue la **plus petite quantité d'information possible**. Mais dès les débuts de l'informatique, les ordinateurs traitaient l'information par **petits paquets de plusieurs bits**, ces "bouchées" (*bite* en anglais) ont donné le mot **byte** (symbole "B") en anglais. Ensuite, lorsque les bytes sont devenus de façon standard des **paquets de 8 bits** est apparu le mot **octet** (symbole "o"), qui aujourd'hui encore est **l'unité de base pour mesurer la quantité d'informations numériques**.

**Remarque :** Byte et octet sont tous les deux utilisés en anglais, et bien que leurs significations soient légèrement différentes, ils sont souvent confondus. De même, la proximité des symboles B pour byte et b pour bit entrainent souvent des confusions, ce qui est très génant car en général 1 B = 8 b = 1 o.


Jusqu'en 1998, on utilisait sans soucis les **multiples** suivants :
- $1 \,ko = 2^{10} \,o = 1024 \,o$
- $1 \,Mo = 2^{20} \,o = 1024 \,ko$
- $1 \,Go = 2^{30} \,o = 1024 \,Mo$
- $1 \,To = 2^{40} \,o = 1024 \,Go$

L'informatique fonctionnant en binaire, il est pratique d'utiliser des puissances de 2 comme multiples et l'approximation $2^{10} \approx 1000$ permettait d'utiliser les préfixes habituels. Mais partout ailleur le préfixe kilo (symbole "k") signifie $\times \, 1000$ et pas $\times \, 1024$. De même, le préfixe méga (symbole "M") signifie $\times \, 10^6$ et pas $\times \, 1,048576\cdot 10^6$ et les préfixes giga (symbole "G") et tera (symbole "T") signifient aussi exactement $\times \, 10^9$ et $\times \, 10^{12}$, pas une approximation de ces puissances de 10.

Pour mettre de l'ordre dans cette particularité informatique, il a été décidé depuis 1998 de créer de **nouveaux préfixes correspondant aux puissances binaires** : le kibioctet (symbole Kio) pour "kilo binaire", le mébioctet (symbole Mio), le gibioctet (symbole "Gio") et le tébioctet (symbole "Tio"). On a donc :
- $1 \,Kio = 2^{10} \,o = 1024 \,o$
- $1 \,Mio = 2^{20} \,o = 1024 \,ko$
- $1 \,Gio = 2^{30} \,o = 1024 \,Mo$
- $1 \,Tio = 2^{40} \,o = 1024 \,Go$

Les **préfixes standards** sont eux utilisés avec leur sens habituel :
- $1 \,ko = 10^{3} \,o = 1000 \,o$
- $1 \,Mo = 10^{6} \,o = 1000 \,ko$
- $1 \,Go = 10^{9} \,o = 1000 \,Mo$
- $1 \,To = 10^{12} \,o = 1000 \,Go$

Malheureusement, il est **facile de définir de nouveaux standards mais il est beaucoup plus compliqué de les faire appliquer**, surtout lorsqu'il faut pour cela modifier les logiciels (et en particulier les systèmes d'exploitation) de milliards de machaines, avec tous les risques que cela comporte.

Finalement, aujourd'hui, les deux notations cohabitent entrainant de **nombreuses confusions**, notamment sur la taille des supports de stockage. Si vous achetez un disque de $1 \, To = 10^{12}\, o$, ne soyez pas surpris qu'une fois installé, l'ordinateur indique que vous disposez de $10^{12}\,/\,2^{40}=0.91 \,To$. C'est juste que l'ordinateur devrait plutôt écrire $0.91 \, Tio$.
